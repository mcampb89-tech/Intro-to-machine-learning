#Numpy

```python
import numpy as np

np.random.seed(100)
ar_1d = np.random.rand(25)

ar_2d=ar_1d.reshape(5,5)

squared_matrix = ar_2d ** 2

small = squared_matrix [ squared_matrix < 1e-2]

indices = np.where(squared_matrix < 1e-2)

print("original 1d", ar_1d)
print("5x5",ar_2d)
print("squared matrix", squared_matrix)
print("small values", small)
print("indices", indices)
```
